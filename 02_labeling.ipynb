# 🏨 Hotel Curator — Labeling Pipeline

**Goal**: Apply weak supervision (keyword-based rules) to label all clauses  
across 5 hotel attributes: Cleanliness, Staff Service, Wi-Fi, Noise Level, Location.

**Label schema**: For each clause × attribute → `positive` / `negative` / `not_mentioned`
